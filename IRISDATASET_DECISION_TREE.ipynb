```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

#baray har gol 4 ta feature hast sepal mishe andaze va andaze hay mokhtalef ra be ma neshan midahad

```python
data_iris=datasets.load_iris()
data_iris
```

```python
dataframe=pd.DataFrame(data_iris.data,columns=data_iris.feature_names)
dataframe
```

```python
y=data_iris.target
```

#initial decision tree

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTreeClassifier
```

```python
iris_tree=DecisionTreeClassifier()
iris_tree.fit(dataframe,y)
iris_tree.predict([
                   [5.1,3.5,1.4,0.2]
])
```
