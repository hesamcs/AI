# define power function

```python
def power(instance):
    energy=0
    for i in instance:
        i=int(i)
        energy=energy+(pow(i,2))
    return energy/(len(instance))
```

normal probability density function

```python
import math
def normal(pw,mn,sd):
  help=sd*(math.sqrt(2*math.pi))
  help=1/(help)
  help2=(pw-mn)/(sd)
  help2=pow(help2,2)
  help2=(-0.5)*help2
  help2=math.exp(help2)
  return help*help2
normal=np.vectorize(normal)
```

# 2.b sign fuction algorithm zero crossing rate 

```python
def sgn(x):
  if x>=0:
    return 1
  else:
    return -1


def sgn_finall(instance):
  result=0
  for i in range(1,len(instance)):
    result+=abs(sgn(instance[i])-sgn(instance[i-1]))
  return result
```
