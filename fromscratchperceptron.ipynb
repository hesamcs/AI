```python
import numpy as np
import random
```

```python
bias=1
x=[
   [1,1,1],
   [1,0,1],          # problem of sanp in 2 dimention
   [0,1,1],
   [0.5,0.5,1],
   [0.4,0.4,0],
   [0,0,0]
]
# chon 2 ta feature vojod darad bayad 2 ta weight bashe va bias bayad yek weight bayad dashte bashad
#bias hich vaght nemitone sefr bashe dar vaghe weight bias sefr ast
weight=[
        random.random(),   #first feature
        random.random(),   # secound feature
        random.random()    # bias
]
weight
```

```python
def activation_func(number): #heaviside activation function
  if number>0:
    return 1
  return 0
```

```python
def noron(instance):
  x0=instance[0]
  x1=instance[1]
  y=instance[2] #response variable
  
  output=(x0*weight[0])+(x1*weight[1])+(bias*weight[2])
  output=activation_func(output)
  error=y-output
  print(error)
  # back propagation
  weight[0]=weight[0]+(error * x0)
  weight[1]=weight[1]+(error * x1)
  weight[2]=weight[2]+(error * bias)
  print(weight)
```

```python
for i in range(100):
  for j in x:
    # print(j) # har epoch
    noron(j)
  print("epoch: ",i)

```
