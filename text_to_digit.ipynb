```python
import networkx as nx

one_to_ten=["صفر", "یک", "دو", "سه","چهار", "پنج", "شش", "هفت", "هشت", "نه", "ده", "یازده"
             , "دوازده", "سیزده", "چهارده", "پانزده", "شانزده", "هفده", "هجده", "نوزده"]
twenty_to_90 = [ "بیست","سی", "چهل", "پنجاه", "شصت", "هفتاد", "هشتاد", "نود"]
hundred_to_900 = [ "صد", "دویست", "سیصد", "چهارصد", "پانصد", "ششصد", "هفتصد", "هشتصد", "نهصد"]
thousand_to_9000 = ["هزار","دو هزار","سه هزار","چهار هزار","پنج هزار","شش هزار","هفت هزار","هشت هزار","نه هزار","ده هزار","یازده هزار","دوازده هزار","سیزده هزار","چهارده هزار","پانزده هزار","شانزده هزار","هفده هزار","هجده هزار","نوزده هزار"]
twenty_thousand_to_90000=[ "بیست هزار","سی هزار", "چهل هزار", "پنجاه هزار", "شصت هزار", "هفتاد هزار", "هشتاد هزار", "نود هزار"]
hundred_thousand_to_900000=[ "صد هزار", "دویست هزار", "سیصد هزار", "چهارصد هزار", "پانصد هزار", "ششصد هزار", "هفتصد هزار", "هشتصد هزار", "نهصد هزار"]



dct={}
graph_1=nx.Graph()
for i in one_to_ten:graph_1.add_node(i)
h=0
for word in one_to_ten:graph_1.add_edge(word,word,weight=h);h=h+1
for word in one_to_ten:weg=graph_1.get_edge_data(word,word);dct[word]=weg["weight"]

graph_2=nx.Graph()
for i in twenty_to_90:graph_2.add_node(i)
h=2
for word in twenty_to_90:graph_2.add_edge(word,word,weight=h*10);h=h+1
for word in twenty_to_90:weg=graph_2.get_edge_data(word,word);dct[word]=weg["weight"]

    

graph_3=nx.Graph()
for i in hundred_to_900:graph_3.add_node(i)
h=1
for word in hundred_to_900:graph_3.add_edge(word,word,weight=h*100);h=h+1
for word in hundred_to_900:weg=graph_3.get_edge_data(word,word);dct[word]=weg["weight"]


graph_4=nx.Graph()
for i in thousand_to_9000:graph_4.add_node(i)
h=1
for word in thousand_to_9000:graph_4.add_edge(word,word,weight=h*1000);h=h+1
for word in thousand_to_9000:weg=graph_4.get_edge_data(word,word);dct[word]=weg["weight"]

graph_5=nx.Graph()
for i in twenty_thousand_to_90000:graph_5.add_node(i)
h=2
for word in twenty_thousand_to_90000:graph_5.add_edge(word,word,weight=h*10000);h=h+1
for word in twenty_thousand_to_90000:weg=graph_5.get_edge_data(word,word);dct[word]=weg["weight"]

graph_6=nx.Graph()
for i in hundred_thousand_to_900000:graph_6.add_node(i)
h=1
for word in hundred_thousand_to_900000:graph_6.add_edge(word,word,weight=h*100000);h=h+1
for word in hundred_thousand_to_900000:weg=graph_6.get_edge_data(word,word);dct[word]=weg["weight"]

dct["یک میلیون"]=1000000
dct
```

```python
def text_to_digit(text):

  #tokenization
  text=text.split(" و ")
  counter=0

  if ("هزار") in text:
    #if "هزار" in list except index 0
    for sub_of_text in text:
  
      check=("هزار") in sub_of_text
      if check ==True:
        break
      counter+=1


    #add ("هزار") to back list
    if counter >0:
      for i in range(counter):
        check_m=("یک میلیون") in text[i]
        if check_m == False:
          text[i]=text[i]+" هزار"

      
  #generate digit with data base
  digit=0
  for i in text:
    digit+=dct[i]
  return digit

```
